```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Непрерывное распределение

Сгенерируем выборку объёма 1000 из экспонициального распределения ( $loc=0$ и $scale=1$):

```python
expon_rv = sts.expon(0, 1)
sample = expon_rv.rvs(1000)
```

Построим теоретическую и эмпирическую функици распределения:


```python
x = np.linspace(0,5,1000)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf, label='theoretical CDF')

# для построения ECDF используем библиотеку statsmodels
from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label='ECDF')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

Нарисуем гистограмму выборки  и теоретическую функцию плотности:

```python
x = np.linspace(0, 5, 1000)
plt.hist(sample, bins=20, density=True)
plt.plot(x, expon_rv.pdf(x), label = 'probability density')

plt.ylabel('density')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

```python
x = np.linspace(-5, 5, 1000)
plt.plot(x, expon_rv.pdf(x), label = 'probability density', alpha = 0.5)

plt.ylabel('density')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

# Центральная предельная теорема

Сгенерируем 1000 выборок объёма $n = 5, 10, 50$ и $100$ из экспонициального распределения ( $loc=0$ и $scale=1$):


```python
n = 5
Count = 1000

expon_rv = sts.expon(0, 1)
sample = 0

for i in range(n):
  sample = sample + expon_rv.rvs(Count)
sample = sample / n

plt.hist(sample, bins=20, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
# Параметры распределения выборочных средних
Mean_exp = 1
Variance_exp = 1
Std_exp = 1

# Параметры нормального распределения
Mean_norm = Mean_exp
Std_norm = Std_exp / np.sqrt(n)
```

Нарисуем поверх гистограмм плотность соответсвующих нормальных распределений:

```python
n = 5
Count = 1000

# Параметры нормального распределения
Mean_norm = Mean_exp
Std_norm = Std_exp / np.sqrt(n)

expon_rv = sts.expon(0, 1)
norm_rv = sts.norm(Mean_norm, Std_norm)
sample = 0

for i in range(n):
  sample = sample + expon_rv.rvs(Count)
sample = sample / n

plt.hist(sample, bins=20, density=True)

x = np.linspace(0, 2, 100)
plt.plot(x, norm_rv.pdf(x), label = 'theoretical pdf')

plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```
