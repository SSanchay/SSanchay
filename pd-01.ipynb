### Шаг 2. Читаем датасет в переменную chipo

```python
# tsv файл - это tab separated values, чтобы корректно его загрузить, укажи аргумент sep='\t'
path = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo = pd.read_csv(path, sep = '\t')
chipo
```

### Шаг 8. Какое блюдо заказывали больше всего?

```python
popular_name = chipo['item_name'].value_counts().index[0]
popular_name
```

### Шаг 9. Сколько раз заказывали самое популярное блюдо?

```python
chipo[chipo['item_name'] == popular_name]['quantity'].sum()
```

### Шаг 10. Какое блюдо в колонке choice_description заказывали больше всего?

```python
chipo['choice_description']\
    .apply(lambda x: x.replace('[', '').replace(']', '').split(', ') if x == x else x)\
    .explode().value_counts().index[0]
```

### Шаг 11. Сколько суммарно блюд было заказано?

```python
chipo['quantity'].sum()
```

### Шаг 12. Измени тип колонки `price` в `float`

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: x.replace('$', '')).astype(float)
chipo.dtypes
```

### Шаг 13. Сколько денег суммарно было получено?

```python
revenue = chipo['item_price'].sum()
revenue
```

### Шаг 14. Сколько заказов суммарно было сделано?

```python
number_orders = chipo['order_id'].nunique()
number_orders
```

### Шаг 15. Какая средняя цена заказа?

```python
# Решение 1
round(revenue/number_orders, 2)
```

```python
# Решение 2
chipo.groupby('order_id').sum().mean()['item_price']
```
