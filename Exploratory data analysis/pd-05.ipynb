# Заказы в фастфуде Chipotle

### Шаг 1. Импортируем библиотеки.

```python
import pandas as pd
```

### Шаг 2. Прочитаем датасет в переменную chipo

```python
path = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
# уже знаете откуда взять данные?)
```

```python
chipo = pd.read_csv(path, sep='\t')
chipo
```

### Шаг 3. Сколько блюд стоят больше $10.00?

```python
chipo['item_name'][chipo['item_price'].apply(lambda x: float(x.replace('$', ''))) > 10].nunique()
```

### Шаг 4. Сколько стоит каждое блюдо?
###### выведите датафрейм с колонками item_name и item_price

```python
chipo[['item_name', 'item_price']].drop_duplicates()
```

### Шаг 5. Сортируем по названию блюда

```python
chipo.sort_values(by = 'item_name')
```

### Шаг 6. Сколько раз заказывали самое дорогое блюдо?

```python
chipo[chipo['item_price'].apply(lambda x: float(x.replace('$', ''))) \
    == chipo['item_price'].apply(lambda x: float(x.replace('$', ''))).max()]\
        .shape[0]
```

### Шаг 7. Сколько раз заказывали Veggie Salad Bowl?

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl']['order_id'].nunique()
```

### Шаг 8. В скольких заказах присутствует больше одной Canned Soda?

```python
#Количество подряд заказанных напитков за один раз
(chipo[['order_id', 'quantity', 'item_name']][chipo['item_name'] == 'Canned Soda']\
    .groupby('order_id').agg({'quantity': 'sum'})> 1).sum()
```
