### Шаг 1. Импортируем библиотеки

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Шаг 2. Создайте датафрейм с 5 элементами, который будет выглядеть так:

```python
df = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
                    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
                    'age': [42, 52, 36, 24, 73],
                    'female': [0, 1, 1, 0, 1],
                    'preTestScore': [4, 24, 31, 2, 3],
                    'postTestScore': [25, 94, 57, 62, 70]})
df
```

### Шаг 3. Постройте scatterplot preTestScore и postTestScore, с размером точки по возрасту

```python
fig, ax = plt.subplots(figsize = (12, 6))

ax.scatter('preTestScore',
            'postTestScore',
            s=df['age']*5,
            data=df)
ax.set(
    xlabel = 'preTestScore',
    ylabel = 'PostTestScore',
    title = 'Test score with size of age'
)
plt.grid()
plt.show()
```

### Шаг 4. Постройте scatterplot preTestScore и postTestScore.
### В этот раз размер должен равен 4.5 \* postTestScore и цветом по полу

```python
sns.set_style('darkgrid')
sns.scatterplot(x = 'preTestScore',
        y = 'postTestScore',
        s = 4.5*df['postTestScore'],
        c = df['female'],
        hue = 'female',
        data=df).\
        set(title ='TestScore with colour of sex');
```
