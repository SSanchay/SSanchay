### Шаг 1. Импортируем библиотеки

```python
import pandas as pd
```

### Шаг 2. Сделаем 3 разных Series длины 100:
1. Случайные числа от 1 до 4
2. Случайные числа от 1 до 3
3. Случайные числа от 10 000 до 30 000

```python
series_1 = pd.Series(range(1, 5)).sample(100, replace=True)
series_2 = pd.Series(range(1, 4)).sample(100, replace=True)
series_3 = pd.Series(range(10000, 30000)).sample(100, replace=True)
```

### Шаг 3. Сделаем датафрейм из этих колонок

```python
df = pd.DataFrame({0: series_1.values, 1: series_2.values, 2: series_3.values})
df
```

### Шаг 4. Назовите колонки bedrs (комнаты), baths (ванные), price_sqr_meter (цена за метр)

```python
df = df.rename(columns={0: 'bedrs', 1: 'baths', 2: 'price_sqr_meter'})
```

```python
df.columns = ['bedrs', 'baths', 'price_sqr_meter']
df
```

### Шаг 5. Сделайте один длинный датафрейм bigcolumn из 3 Series с одной колонкой

```python
bigcolumn = pd.DataFrame(pd.concat((series_1, series_2, series_3)))
bigcolumn
```

### Шаг 6. Кажется, индекс идет только до 99. Проверим?

```python
bigcolumn.index.max() < 100
```
